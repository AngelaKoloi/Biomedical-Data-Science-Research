```python
import pandas as pd

real = pd.read_csv('df_cleaned.csv') 
synthetic = pd.read_csv('balanced_data_lifelines.csv') 
```

```python
def create_combined_heatmaps(real_data, synthetic_data, continuous_vars, categorical_vars, threshold=0.1):
    all_vars = continuous_vars + categorical_vars
    
    # Create a custom correlation matrix with mixed methods
    def compute_mixed_corr(data):
        corr_matrix = pd.DataFrame(index=all_vars, columns=all_vars)
        
        for i in all_vars:
            for j in all_vars:
                if i in continuous_vars and j in continuous_vars:
                    # Continuous-continuous: Pearson
                    corr = data[i].corr(data[j], method='pearson')
                else:
                    # Mixed or categorical-categorical: Spearman
                    corr = data[i].corr(data[j], method='spearman')
                corr_matrix.loc[i, j] = corr
        return corr_matrix.astype(float)
    
    # Compute correlation matrices
    real_corr = compute_mixed_corr(real_data)
    synthetic_corr = compute_mixed_corr(synthetic_data)

    # Calculate similarity percentages (NEW)
    diff_matrix = (real_corr - synthetic_corr).abs()
    total_elements = diff_matrix.size
    similar_elements = (diff_matrix <= threshold).sum().sum()
    
    similarity_percentage = (similar_elements / total_elements) * 100
    difference_percentage = 100 - similarity_percentage
    
    print(f"\nCorrelation Similarity Report:")
    print(f"Similarity (≤{threshold} difference): {similarity_percentage:.1f}%")
    print(f"Difference (> {threshold} difference): {difference_percentage:.1f}%")
    
    # Create mask for triangular matrix
    mask = np.triu(np.ones_like(real_corr, dtype=bool))
    
    # Create figure with custom layout
    fig = plt.figure(figsize=(36, 20))
    
    # Create gridspec with space for colorbar on top
    gs = fig.add_gridspec(2, 2, height_ratios=[0.05, 1], width_ratios=[1, 1],
                         hspace=0.05, wspace=0.2)
    
    # Create axes
    ax1 = fig.add_subplot(gs[1, 0])  # Real data heatmap
    ax2 = fig.add_subplot(gs[1, 1])  # Synthetic data heatmap
    cbar_ax = fig.add_subplot(gs[0, :])  # Colorbar across top
    
    # Common plot settings
    heatmap_kws = {
        'mask': mask,
        'annot': False,
        'fmt': ".2f",
        'cmap': "coolwarm_r",  # Reversed colormap (blue=positive, red=negative)
        'cbar': False,
        'vmin': -1,
        'vmax': 1
    }
    
    # Plot real data heatmap
    sns.heatmap(real_corr, ax=ax1, **heatmap_kws)
    ax1.tick_params(axis='both', which='major', labelsize=18)
    ax1.set_title("Real Data", fontsize=18, pad=20, y=1.02)
    
    # Plot synthetic data heatmap
    im = sns.heatmap(synthetic_corr, ax=ax2, **heatmap_kws)
    ax2.tick_params(axis='both', which='major', labelsize=18)
    ax2.set_title("Synthetic Data", fontsize=18, pad=20, y=1.02)
    
    # Add centered colorbar on top
    cbar = fig.colorbar(im.collections[0], cax=cbar_ax, orientation='horizontal')
    cbar.set_label('Correlation Coefficient', size=18, labelpad=10)
    cbar.ax.tick_params(labelsize=18)
    cbar_ax.xaxis.set_ticks_position('top')
    cbar_ax.xaxis.set_label_position('top')
    
    plt.savefig('/groups/umcg-lifelines/tmp02/projects/ov20_0110/Lifelines/results/real_synthetic_comparison_lifelines.png', dpi=600, bbox_inches='tight')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

continuous_vars = [ ] 
categorical_vars = [ 'Diastolic blood pressure', 'Systolic blood pressure',
    'Age', 'Cholesterol', 'HDL cholesterol', 'LDL cholesterol', 'Triglycerides',
    'Glucose', 'Childhood trauma score',
    'Depressive symptoms score', 'Anxiety symptoms score',  
    'Physically abused by family as a child',
    'Felt hated by family member as a child',
    'Sexually molested as a child',
    'Someone to take to doctor when needed as a child',
    'Felt loved', 
    'Hypertension', 'Smoking status', 'Physical activity', 
'Depressed mood', 'Anhedonia', 'Appetite changes', 'Sleep problems',
    'Psychomotor changes', 'Fatigue', 'Feelings of inadequacy',
    'Cognitive problems', 'Suicidal ideation', 'Anxiety', 'Restlessness',
    'Lack of relaxation',  'Concentration problems',
    'Irritability', 
    'Antidepressant use', 'Diabetes', 'Gender',
    'CVD Family history']
create_combined_heatmaps(real, synthetic, continuous_vars, categorical_vars)
```
